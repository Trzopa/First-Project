# Indicates

## CPI

```python
fed_data = df[df['Year'] >= 1913]

sns.set(style='whitegrid')

plt.figure(figsize=(12, 6))

sns.lineplot(x='Year', y='CPI', data=fed_data)

plt.xticks(range(1910, 2025, 10))
plt.yticks(range(0, 320, 20))
plt.title('Changes in the CPI Index from 1913 to 2023', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('CPI', fontsize=12)

plt.show()
```

Up until the 1970s, the CPI index did not rise dynamically; at some points, there were even slight declines or it remained at the same level. Since the early 1970s, the CPI began to increase drastically year by year, although there were short periods of stabilization. After the year 2000, the CPI index continued to rise, and in recent years, the pace of this growth has accelerated. In 2023, the CPI index reached the highest level in the entire period analyzed.

The main reasons for the rise in the CPI index in the 1970s include the end of the gold-backed U.S. dollar and the rise in oil prices, which led to an increase in the prices of all products.

## Inflation

```python
fed_data = df[df['Year'] >= 1913]

sns.set(style='whitegrid')

plt.figure(figsize=(12, 6))
ax = plt.gca()

x = fed_data['Year']
y = fed_data['Inflation']

scatter = ax.scatter(x, y, c=y, cmap='coolwarm', norm=plt.Normalize(-10, 10))

plt.axhline(0, color='red', linewidth=1, linestyle='--')

ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min(), y.max())

plt.title('Changes in the Inflation index from 1913 to 2023', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Inflation', fontsize=12)

plt.xticks(range(1910, 2025, 5), rotation=45)
plt.yticks(range(-15, 20, 5))

plt.colorbar(scatter, ax=ax, label='Inflation')

plt.show()
```

High inflation, particularly evident in the early 1920s and the early 1980s, is often the result of external economic shocks or drastic changes in monetary policy. For instance, the early 1920s might reflect the aftermath of World War I, which disrupted economies and resources, leading to shortages and higher prices. Similarly, the early 1980s experienced high inflation partly due to high oil prices and monetary policies implemented in response to previous oil crises, which focused on economic recovery at the cost of higher inflation.

Conversely, prolonged periods of low inflation, observed in the second half of the 20th century and the early 21st century, may be associated with more stable economic conditions, effective monetary policy, and widespread globalization. The introduction of new technologies, which reduced production costs and improved logistics, also contributed to price stabilization. Additionally, the growing integration of global markets and the development of the service sector enabled better management of inflationary pressure.

Periods of deflation, particularly evident during the Great Depression, are often the result of massive disruptions in the economy, such as excessive debt, bursting speculative bubbles, drastic cuts in consumer and investment spending, which lead to a decline in demand and prices.

Overall, inflation reflects many factors, including government policies, the state of the global economy, changes in commodity prices, dynamics of supply and demand, and the expectations of consumers and investors. Each historical period is characterized by a unique set of circumstances that collectively influence the level of inflation.

## Economic indicators in the years 2004-2023

### Inflation

```python
data_subset = data_2023_down.head(20)
sns.set(style='whitegrid')

plt.figure(figsize=(12, 6))
sns.regplot(x='Year', y='Inflation', data=data_subset, marker='o', scatter_kws={'s': 50}, line_kws={'color': 'blue', 'linewidth': 2})

plt.title('Changes in the Inflation Index from 2004 to 2023', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Inflation', fontsize=12)

plt.xticks(range(2004, 2024, 1), rotation=45)

plt.show()
```

The chart showing changes in the inflation index from 2004 to 2023 reveals a general upward trend, but with significant details in the context of major economic events. In 2008, while it might seem that the financial crisis could have caused an increase in inflation, the main issue was actually a crash in the real estate market and a credit crisis, which did not necessarily translate into higher inflation. During this time, price stagnation was observed instead, as falling economic activity and the credit crisis suppressed consumer and investment demand.
However, a significant rise in inflation is noticeable in the years 2021-2023, which can be linked to the long-term effects of the COVID-19 pandemic. The pandemic caused global disruptions in supply chains, which, combined with economic recovery following the initial lockdowns, led to increased demand alongside limited supply. This, in turn, translated into a rise in prices, as clearly visible on the chart.

Analyzing the chart helps understand how various events can impact inflation in different periods. While the financial crisis in 2008 did not bring about a significant increase in inflation due to decreased demand and market collapse, the pandemic and related supply chain issues caused considerable inflation in subsequent years. This highlights the complexity of the relationship between economic conditions and changes in price levels.

### Unemployment rate

```python
data_subset04_23 = sort_93_23[(sort_93_23['yr'] >= 2004) & (sort_93_23['yr'] <= 2023)]

sns.set(style='whitegrid')

plt.figure(figsize=(12, 6))
sns.lineplot(x='yr', y='Unemployment rate', data=data_subset04_23)

plt.xticks(range(2004,2023,1), rotation=45)
plt.title('Unemployment in USA 2004-2023')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)

plt.show()
```

Analyzing the unemployment rate chart in the USA from 2004-2023, several key trends and events can be observed that have influenced the labor market.

The chart shows that the unemployment rate was relatively low before the 2008 financial crisis, where it was below 6%. The increase in unemployment that occurred in 2009 is directly linked to the global financial crisis and the real estate market crisis in the USA, which led to mass layoffs and company bankruptcies, causing unemployment to spike to nearly 10%.

Subsequently, the unemployment rate began to gradually decrease, reflecting the slow economic recovery after the crisis, supported by government interventions and the quantitative easing policy conducted by the Federal Reserve. This recovery continued until 2020, when the unemployment rate suddenly rose again, attributable to the impact of the COVID-19 pandemic. Lockdowns and restrictions implemented to combat the pandemic led to the closure of many businesses and a sharp increase in unemployment.

After a brief spike in 2020, the unemployment rate quickly dropped, which may be a result of markets adapting to new conditions, dynamic reopening of the economy, and extensive government support programs for businesses and workers. Significant declines in the unemployment rate in subsequent years indicate a relatively quick return to stabilization, although the economic environment remained full of uncertainties related to the ongoing effects of the pandemic.

### Gross domestic product per capita, constant prices

```python
data_subset04_23 = sort_93_23[(sort_93_23['yr'] >= 2004) & (sort_93_23['yr'] <= 2023)]

average_gdp_per_capita = data_subset04_23.groupby('yr')['Gross domestic product per capita, constant prices'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.set(style='whitegrid')

sns.lineplot(x='yr', y='Gross domestic product per capita, constant prices', data=average_gdp_per_capita)

plt.xticks(range(2004,2023,1), rotation=45)
plt.title('Gross domestic product per capita, constant prices in USA 2004-2023')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Gross domestic product per capita, constant prices (USD)', fontsize=10)

plt.show()
```

Analyzing the Gross Domestic Product per capita at constant prices in the USA for the years 2004-2023, we see an overall upward trend, indicating an improvement in economic well-being. The significant drop in 2008-2009 reflects the impact of the global financial crisis, but the rapid recovery after this period and stable growth until 2019 suggest the effectiveness of policies stabilizing the economy. The recent increase post-2020 may be the result of a rebound after the COVID-19 pandemic, showing the dynamic response of the economy to external challenges and its ability to quickly regenerate. This chart demonstrates the resilience and adaptive capacity of the US economy, which despite global crises, can find ways to grow and stabilize.

### Gross domestic product based on purchasing-power-parity

```python
data_subset04_23 = sort_93_23[(sort_93_23['yr'] >= 2004) & (sort_93_23['yr'] <= 2023)]
average_gdp_per_capita = data_subset04_23.groupby('yr')['Gross domestic product based on purchasing-power-parity (PPP) share of world total'].mean().reset_index()

sns.set(style='whitegrid')

plt.figure(figsize=(12, 6))
sns.lineplot(x='yr', y='Gross domestic product based on purchasing-power-parity (PPP) share of world total', data=average_gdp_per_capita)

plt.xticks(range(2004,2023,1), rotation=45)
plt.title('Gross domestic product based on purchasing-power-parity in USA 2004-2023', fontsize=10)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Gross domestic product based on purchasing-power-parity (%)', fontsize=10)

plt.show()
```

The chart depicting the share of the USA's GDP based on purchasing power parity (PPP) in global GDP from 2004-2023 shows a general downward trend. Starting at nearly 19% in 2004, this share has systematically decreased, reaching below 16% in 2023. This trend may result from faster economic growth in other countries, particularly in the developing economies of Asia, which relatively reduces the share of the American economy in global GDP. This phenomenon may also be a result of the global redistribution of production and investment, shifting the economic center of gravity from the United States to other regions of the world.

### Volume of Exports and Imports of Goods and Services

```python
data_subset_exports = sort_93_23[(sort_93_23['yr'] >= 2004) & (sort_93_23['yr'] <= 2023)]
average_exports = data_subset_exports.groupby('yr')['Volume of exports of goods and services'].mean().reset_index()

data_subset_imports = sort_93_23[(sort_93_23['yr'] >= 2004) & (sort_93_23['yr'] <= 2023)]
average_imports = data_subset_imports.groupby('yr')['Volume of imports of goods and services'].mean().reset_index()

sns.set(style='whitegrid')

plt.figure(figsize=(10, 6))
plt.axhline(0, color='red', linewidth=2, linestyle='--')

sns.lineplot(x='yr', y='Volume of exports of goods and services', data=average_exports, label='Exports')

sns.lineplot(x='yr', y='Volume of imports of goods and services', data=average_imports, label='Imports')

plt.xticks(range(2004, 2024, 1), rotation=45)
plt.yticks(range(-20, 16, 2))
plt.title('Volume of Exports and Imports of Goods and Services in USA 2004-2023', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Volume of Exports and Imports (%)', fontsize=12)

plt.legend()

plt.show()
```

Analyzing the chart depicting the volume of exports and imports of goods and services in the USA from 2004-2023, several significant trends and events can be observed that have impacted international trade.

The chart shows that both exports and imports are characterized by significant volatility throughout the period. Sharp declines and increases may be correlated with key economic and political events. For example, the steep drop in both indicators in 2009 reflects the global financial crisis, which resulted in a significant economic slowdown worldwide, limiting both the demand for exported goods and services and the capacity for imports.

Subsequently, we notice a gradual recovery in the years following the crisis, indicating an improvement in economic conditions and a resumption of trade activities. However, the most dramatic changes are observed in 2020 and 2021, which are direct consequences of the COVID-19 pandemic. This pandemic caused global disruptions in supply chains and significant restrictions on international trade, explaining the sharp decline in both indicators in 2020.

Interestingly, in 2021, despite the continuation of the pandemic, we observe a rapid increase in both exports and imports, which may indicate a quick adaptation of companies to new conditions and the resumption of international trade, possibly thanks to the application of new technologies and transportation methods, as well as government stimulus programs aimed at rebuilding the economy.

Overall, the chart shows how global events, such as financial crises and pandemics, have a direct impact on international trade. Reactions to these events, both nationally and globally, determine the speed of recovery and the markets’ adaptation to changing conditions.

```python
data_subset = data_1995_down.head(30)
sns.set(style='whitegrid')

plt.figure(figsize=(12, 6))
sns.regplot(x='Year', y='Inflation', data=data_subset, marker='o', scatter_kws={'s': 50}, line_kws={'color': 'blue', 'linewidth': 2})

plt.title('Changes in the Inflation Index from 1964 to 1995', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Inflation', fontsize=12)

plt.xticks(range(1964, 1996, 2), rotation=45)
plt.show()
```

Analyzing the inflation index changes from 1964 to 1995, one can notice a general downward trend in inflation during this period. Initially, in the 1960s, inflation remained at a relatively high level, which can be attributed to significant government spending associated with social programs and military interventions, such as the Vietnam War. High government spending contributed to an increase in demand, leading to rising prices.

In the 1970s, inflation was particularly volatile and reached its peaks, mainly due to two oil crises in 1973 and 1979. The significant increase in oil prices, resulting from the embargo imposed by OPEC countries and later supply disruptions, triggered an inflationary wave worldwide, including in the USA. Inflation during this period was also supported by monetary policy, which initially was not restrictive enough to control rising prices.

From the late 1970s and early 1980s, there was a shift in monetary policy, mainly due to decisions taken by the Federal Reserve under the leadership of Paul Volcker. Restrictive monetary measures were introduced, including significant interest rate hikes, aimed at curbing inflation. These actions were effective, although they also had short-term negative effects in the form of a recession in the early 1980s.

In the 1980s and 1990s, inflation was much lower and more stable, which can be attributed to the continuation of restrictive monetary policy and stabilization in commodity markets. Globalization and technological progress also contributed to maintaining low inflation by reducing production costs and increasing efficiency.

In summary, changes in inflation from 1964 to 1995 were the result of complex interactions between fiscal policy, monetary policy, and global economic events such as oil crises and changes in the structure of the global economy. The early years were characterized by high inflation due to government spending and external shocks, while the later years brought stabilization and a decrease in inflation thanks to more effective monetary policies and global changes.

Analyzing the chart of inflation index changes from 1935 to 1965, it can be noted that the overall trend in inflation was slightly downward. In the 1930s and 1940s, inflation was relatively high, stemming from a series of significant historical and economic events.

In the 1930s, the world grappled with the Great Depression, which initially triggered deflation, but subsequent government policies, such as the New Deal in the USA, aimed to revitalize the economy through increased public spending. This contributed to an increase in inflation in the late 1930s and early 1940s.

Another significant factor affecting inflation was the period of World War II. During the war, government spending on armaments and other war-related needs was enormous, significantly increasing demand in the economy. After the war, in the 1940s, high inflation was the result of both the decompression of the economy after years of war and a sudden surge in consumer demand as people began to spend their savings accumulated during the war.

In the 1950s and 1960s, inflation gradually stabilized and decreased. This was due to a more stable economic situation, technological development, increased productivity, and more effective monetary policies conducted by central banks. Economic growth became more balanced, and the global economy began to integrate more fully, which also contributed to greater price stability.

The chart shows that although inflation was variable and notably higher in periods directly related to the war and its aftermath, the overall trend towards stabilization was evident from the 1950s. During this period, economic policy became more focused on controlling inflation and stabilizing the economy, which brought visible results in the form of lower and more stable inflation rates in the following decades.

After the war ended, in the 1920s, inflation began to decline. These years were a period of economic revival known as the "Roaring Twenties." Industrial production growth, technological development, and increased consumption contributed to an initial rise in inflation; however, the economy began to stabilize, and inflation decreased.

The most significant changes in inflation during the analyzed period occurred during the Great Depression from 1929 to 1933. During this time, inflation sharply decreased and turned into deflation. Deflation was a result of mass bankruptcies, a dramatic drop in demand, falling prices of goods and services, and high unemployment. Banks failed, credit was unavailable, and consumers and businesses reduced their spending, causing further price declines.

Concluding from the chart, the period from 1913 to 1935 was a time of significant inflation fluctuations, mainly associated with events such as World War I, the economic boom of the 1920s, and the Great Depression. These events had a crucial impact on the price level, shaping the dynamic economic landscape of that era.
